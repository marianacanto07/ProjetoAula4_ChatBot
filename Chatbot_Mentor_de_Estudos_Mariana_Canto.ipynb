**Nome do Projeto:** Chatbot: Seu mentor de estudos!

O Chatbot mentor de estudos é uma **ferramenta interativa projetada para auxiliar estudantes de maneira dinâmica e adaptativa**. Utilizando o poder da inteligência artificial, o chatbot não só responde a perguntas específicas, mas também analisa o progresso do aluno e pode sugerir **recursos personalizados, como vídeos, artigos e quizzes**, para reforçar o aprendizado. Além disso, o bot é capaz de detectar o nível de compreensão e o estado emocional do usuário através de análises de sentimentos, ajustando sua comunicação para manter o estudante motivado e engajado.

O Chatbot Mentor de Estudos oferece um **apoio contínuo e personalizado aos estudantes**, incentivando um aprendizado mais profundo e eficaz. Ele serve como um companheiro virtual, que acompanha o estudante durante sua jornada educacional, fazendo ajustes e recomendações que ajudam a superar dificuldades e a **maximizar o potencial de aprendizado**.

```python
# Configuração da API do Google GenerativeAI

import google.generativeai as genai
from google.colab import userdata
GOOGLE_API_KEY= userdata.get('SECRET_KEY')

genai.configure(api_key=GOOGLE_API_KEY)
```

```python
# Criação do modelo generativo com configurações avançadas
model = genai.GenerativeModel(
    model_name='gemini-1.0-pro',
    generation_config={
        "candidate_count": 1,
        "temperature": 0.7,
    },
    safety_settings={
        'HATE': 'HIGH',
        'HARASSMENT': 'HIGH',
        'SEXUAL': 'HIGH',
        'DANGEROUS': 'HIGH'
    },
)
```

```python
# Função para melhorar a visualização das respostas do chatbot
def to_markdown(text):
    text = text.replace('•', '  *')
    return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
# Inicialização do chat e interface de usuário
chat = model.start_chat(history=[])
print("Bem-vindo ao Chatbot Mentor de Estudos! Digite sua pergunta ou 'fim' para sair.")

prompt = input('Esperando sua pergunta: ')
while prompt.lower() != "fim":
    response = chat.send_message(prompt)
    display(to_markdown(f'**Chatbot**: {response.text}'))
    prompt = input('Esperando sua pergunta: ')

# Ao final, exibe o histórico completo de interação
from IPython.display import display, Markdown
import textwrap

print("Histórico de chat:")
for message in chat.history:
    display(to_markdown(f'**{message.role}**: {message.parts[0].text}'))
    print('-------------------------------------------')
```
